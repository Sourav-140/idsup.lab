```python
import matplotlib.pyplot as plt

# Dictionary of salaries and tenures
data = {
    0.7: 48000.0,
    1.9: 480000,
    2.5: 60000.0,
    42: 630000,
    6: 760000,
    6.5: 690000,
    7.5: 76000.0,
    8.1: 880000,
    8.7: 83000,
    10: 830000
}

# Extracting tenures and salaries from the dictionary
tenures = list(data.keys())
salaries = list(data.values())

# Plotting the scatterplot
plt.scatter(tenures, salaries)

# Adding labels and title
plt.xlabel('Tenures')
plt.ylabel('Salaries')
plt.title('Employee Salaries vs. Tenures')

# Displaying the plot
plt.show()
```

```python
#calculate slope and intercept

import numpy as np

# Initialize the data
x = np.array([0, 1, 2, 3, 4, 5, 6])
y = np.array([5, 25, 45, 65, 85, 105, 125])

# Initialize the parameters
learning_rate = 0.01
num_iterations = 100
batch_size = 2

# Initialize the slope and intercept
m = 0
c = 0

# Perform minibatch gradient descent
for _ in range(num_iterations):
    # Randomly select a minibatch
    indices = np.random.choice(len(x), size=batch_size, replace=False)
    x_batch = x[indices]
    y_batch = y[indices]

    # Compute the predicted y values
    y_pred = m * x_batch + c

    # Compute the gradients
    m_gradient = (-2/batch_size) * np.sum(x_batch * (y_batch - y_pred))
    c_gradient = (-2/batch_size) * np.sum(y_batch - y_pred)

    # Update the parameters
    m -= learning_rate * m_gradient
    c -= learning_rate * c_gradient

# Print the final slope and intercept
print("Slope (m):", m)
print("Intercept (c):", c)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 500)  # Generate 500 evenly spaced points between -10 and 10
y1 = x * np.sin(x)
y2 = x**2 * np.sin(x)
y3 = x**3 * np.sin(x)
y4 = x**4 * np.sin(x)

plt.plot(x, y1, label='x*sin(x)')
plt.plot(x, y2, label='x^2*sin(x)')
plt.plot(x, y3, label='x^3*sin(x)')
plt.plot(x, y4, label='x^4*sin(x)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of x*sin(x), x^2*sin(x), x^3*sin(x), and x^4*sin(x)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#plotpdf

import numpy as np
import matplotlib.pyplot as plt

# Define the parameters for the normal distribution
mean = 500
std_dev = 11.18

# Generate x values from mean-4*std_dev to mean+4*std_dev
x = np.linspace(mean - 4 * std_dev, mean + 4 * std_dev, 100)

# Calculate the corresponding y values using the normal distribution formula
y = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)

# Create the plot
plt.plot(x, y, linewidth=2)
plt.title('PDF of Normal Distribution')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)

# Show the plot
plt.show()
```
